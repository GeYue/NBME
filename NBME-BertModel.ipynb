```python
import os
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

DATA_PATH = "./data/"
```

```python
trainDF = pd.read_csv(DATA_PATH + "train.csv")
featureDF = pd.read_csv(DATA_PATH + "features.csv")
patientDF = pd.read_csv(DATA_PATH + "patient_notes.csv")
```

```python
def PatientNum2FmtStr (value):
    return "%05d" % value
def FeatureNum2FmtStr (value):
    return "%03d" % value

trainDF['pn_num'] = trainDF['pn_num'].apply(PatientNum2FmtStr)
trainDF['feature_num'] = trainDF['feature_num'].apply(FeatureNum2FmtStr)
featureDF['feature_num'] = featureDF['feature_num'].apply(FeatureNum2FmtStr)
patientDF['pn_num'] = patientDF['pn_num'].apply(PatientNum2FmtStr)
#patientDF['pn_history'] = patientDF['pn_history'].astype(str)
```
